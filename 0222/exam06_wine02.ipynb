```python
# winequality-red.csv 읽어서 출력
import pandas as pd
red_df = pd.read_csv('winequality-red.csv',sep=';', header=0, engine='python')  # sep=';' : 세미콜론으로 구분해서 데이터를 읽어와라
print(red_df)

white_df = pd.read_csv('winequality-white.csv',sep=';', header=0, engine='python')
print(white_df)
```

```python
# winequality-red2.csv  / winequality-white2.csv 내보내기
red_df.to_csv('winequality-red2.csv', encoding='utf-8-sig', index=False)
white_df.to_csv('winequality-white2.csv', encoding='utf-8-sig', index=False)
```

```python
# 두 파일을 합치기 전에 red파일에 있는 것과 white파일에 있는 것 구분해줌
red_df.insert(0, column='type', value='red')  # 첫번째 위치에 값이 red인 type컬럼이 insert됨
```

```python
wine = pd.concat([red_df, white_df])  # concat() : 입력한 데이터의 값을 합쳐줌
```

```python
# 컬럼의 이름에서 공백을 언더바(_)로 변경
wine.columns = wine.columns.str.replace(' ', '_')
```

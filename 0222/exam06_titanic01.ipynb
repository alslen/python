```python
def category_age(x):
    if x<10:
        return 0
    elif x<20:
        return 1
    elif x<30:
        return 2
    elif x<40:
        return 3
    elif x<50:
        return 4
    elif x<60:
        return 5
    elif x<70:
        return 6
    else:
        return 7
    
```

```python
titanic['age2'] = titanic['age'].apply(category_age)  # 각각의 항목에 대해서 값을 가져와서 category_age실행 -> age2컬럼에 값을 추가
titanic['sex2'] = titanic['sex'].map({'male':1, 'female':0})  # sex컬럼에서 male이면 1 female이면 0을 sex2에 추가
titanic['family'] = titanic['sibsp']+titanic['parch']+1  # 1 : 자기자신

# 히스토그램 그림
heatmap_data = titanic[['survived', 'sex2', 'age2', 'family', 'pclass', 'fare']]
colormap = plt.cm.RdBu
sns.heatmap(heatmap_data.astype(float).corr(), linewidths=0.1, vmax=1.0, square=True, cmap=colormap, linecolor='white', annot=True, annot_kws={'size':10})
```
